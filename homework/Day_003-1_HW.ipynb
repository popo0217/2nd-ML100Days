```python
import pandas as pd
import numpy as np
np.floor(np.random.rand(25) * np.random.rand(25))
```

```python
# set 24 countries with random population
# population uses two random.rand to make it more realistic and rounded to integer
data = {'國家': ['Afhganistan', 'Brazil', 'Canada', 'Denmark', 'Ethiopia', 'Finland', 'Germany', 'Haiti', 'Iceland', 'Jamaica', 
               'Kenya', 'Liechtenstein', 'Monaco', 'Netherlands', 'Oman', 'Portugal', 'Qatar', 'Russia', 'Sweden', 'Taiwan',
               'Uruguay', 'Vietnam', 'Yemen', 'Zimbabwe'],
        '人口': (np.random.rand(24) * np.random.rand(24) * 1000000000).astype(int)}
data = pd.DataFrame(data)
data
```

```python
# the country with most population
data[data['人口'] == data['人口'].max()]
```
